```python
dataset['% Iron Feed']=pd.to_numeric(dataset['% Iron Feed'],errors='coerce')
dataset['% Silica Feed']=pd.to_numeric(dataset['% Silica Feed'],errors='coerce')
dataset['Starch Flow']=pd.to_numeric(dataset['Starch Flow'],errors='coerce')
dataset['Amina Flow']=pd.to_numeric(dataset['Amina Flow'],errors='coerce')
dataset['Ore Pulp Flow']=pd.to_numeric(dataset['Ore Pulp Flow'],errors='coerce')
dataset['Ore Pulp pH']=pd.to_numeric(dataset['Ore Pulp pH'],errors='coerce')
dataset['Ore Pulp Density']=pd.to_numeric(dataset['Ore Pulp Density'],errors='coerce')
dataset['Flotation Column 01 Air Flow']=pd.to_numeric(dataset['Flotation Column 01 Air Flow'],errors='coerce')
dataset['Flotation Column 02 Air Flow']=pd.to_numeric(dataset['Flotation Column 02 Air Flow'],errors='coerce')
dataset['Flotation Column 03 Air Flow']=pd.to_numeric(dataset['Flotation Column 03 Air Flow'],errors='coerce')
dataset['Flotation Column 04 Air Flow']=pd.to_numeric(dataset['Flotation Column 04 Air Flow'],errors='coerce')
dataset['Flotation Column 05 Air Flow']=pd.to_numeric(dataset['Flotation Column 05 Air Flow'],errors='coerce')
dataset['Flotation Column 06 Air Flow']=pd.to_numeric(dataset['Flotation Column 06 Air Flow'],errors='coerce')
dataset['Flotation Column 07 Air Flow']=pd.to_numeric(dataset['Flotation Column 07 Air Flow'],errors='coerce')
dataset['Flotation Column 01 Level']=pd.to_numeric(dataset['Flotation Column 01 Level'],errors='coerce')
dataset['Flotation Column 02 Level']=pd.to_numeric(dataset['Flotation Column 02 Level'],errors='coerce')
dataset['Flotation Column 03 Level']=pd.to_numeric(dataset['Flotation Column 03 Level'],errors='coerce')
dataset['Flotation Column 04 Level']=pd.to_numeric(dataset['Flotation Column 04 Level'],errors='coerce')
dataset['Flotation Column 05 Level']=pd.to_numeric(dataset['Flotation Column 05 Level'],errors='coerce')
dataset['Flotation Column 06 Level']=pd.to_numeric(dataset['Flotation Column 06 Level'],errors='coerce')
dataset['Flotation Column 07 Level']=pd.to_numeric(dataset['Flotation Column 07 Level'],errors='coerce')
dataset['% Iron Concentrate']=pd.to_numeric(dataset['% Iron Concentrate'],errors='coerce')
dataset['% Silica Concentrate']=pd.to_numeric(dataset['% Silica Concentrate'],errors='coerce')

```

```python
dataset.corr()
```

```python
plt.figure(figsize=(30, 25))
p = sns.heatmap(dataset.corr(), annot=True)
```

```python
silic_corr = dataset.corr()['% Silica Concentrate']
silic_corr = abs(silic_corr).sort_values()
silic_corr
```

```python
drop_index= silic_corr.index[:5].tolist()
```

Above plot shows the correaltions between the features. From the list we can find out the features which affects the % Silica Concentrate the most and discard the least important ones

```python
dataset= dataset.drop(drop_index, axis=1)
```

```python
dataset.head()
```

SEPARATING DEPENDENT AND INDEPENDENT VARIABLES

```python
target = '% Silica Concentrate'

y = dataset[target]
```

```python
x = dataset.drop([target], axis=1).values
```

```python
x.shape
```

```python
y.shape
```

# #Splitting Train and Test Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Building Model

1)LINEAR REGRESSION

from sklearn.linear_model import LinearRegression

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)#predicted output
```

```python
y_test#actual output
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,lr.predict(x_test))#testing accuracy
```

From this we observed LinearRegression has given accuracy of 67% which is not much,so we try another model for better accuracy

2)Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures()
```

```python
x_poly=poly.fit_transform(x)
```

```python
x_poly
```

```python
x_poly.shape
```

```python
lr.fit(x_poly,y)
```

```python
y_pred=lr.predict(x_poly)
```

```python
r2_score(y,y_pred)
```

From this we observed PolyRegression has given accuracy of 72% which is not much,so we try another model for better accuracy

3)Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor(random_state=0)
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred2=dt.predict(x_test)
```

```python
y_pred2
```

```python
y_test
```

```python
r2_score(y_test,y_pred2)
```

We obtained 99.5% accuracy which is really high.Hence we can consider this model to be accurate and implement it.

```python
plt.figure()
plt.scatter(x=x_test[:,1],y=y_test,color="blue")
plt.scatter(x=x_test[:,1],y=y_pred2,color="red")  
plt.show()
```
